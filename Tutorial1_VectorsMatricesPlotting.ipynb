### Decompose a matrix M into its eigenvectors and eigenvalues

```python
w,v = np.linalg.eig(A)
print(w) # each element of this vector is an eigenvalue
print(v) # each column of this matrix V is eigenvector
```

### Now let us do some simple verification of eigenvalues and eigenvectors

```python
v1=v[:,0] # extract the first eigen vector
print(v1)
```

```python
print(w[0]*v1) # multiply the first eigenvector with the first eigenvalue
```

```python
print(np.matmul(A,v1)) # multiply the matrix A with its first eigenvector
```

### The above two rows show that $Av = \lambda v$ where $v$ is one of the eigenvectors and $\lambda$ is the corresponding eigenvalue of A.

```python
print(np.linalg.det(A)) # This is the determinant using det() function
print(w[0]*w[1]*w[2]) # This is the determinant using multiplication of eigenvalues.
```

### det(A) = $\lambda_1 \times \lambda_2 \times \lambda_3$. The above results is correct within machine precision.

### Matrix multiplication

```python
A=np.array([[1,2,3],
   [4,3,7],
   [7,2,6]])
B=np.array([[4,6,1],
   [9,5,2],
   [11,3,4]])
C=np.array([[ 6,10,7],
 [17,11,16],
 [25,7,16]])
D=np.matmul(A,np.matmul(B,C)) # D = ABC
print(D)
print(np.trace(D)) # And then determine trace of D
```

```python
E=np.matmul(C,np.matmul(A,B)) # E = CAB
print(E)
print(np.trace(E)) # determine trace of E
```

```python
F=np.matmul(B,np.matmul(C,A)) # F = BCA
print(F)
print(np.trace(F)) # trace of F
```

### The above shows that trace(ABC) = trace(CAB) = trace(BCA)

# Plotting a vector

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a single vector
U = np.array([1,0])
V = np.array([1,1])
W = np.array([2,1])

# Create the plot
fig, ax = plt.subplots()

# Add the vector V to the plot
ax.quiver(0, 0, U[0], U[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(1, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='g')
ax.quiver(0, 0, W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')

# Set the x-limits and y-limits of the plot
ax.set_xlim([-1, 2])
ax.set_ylim([-1, 2])

# Show the plot along with the grid
plt.grid(visible='yes')
plt.show()
```

### In Matplotlib 'ax' is an instance of the axes that contain important information about each plot, such as its x- and y-axis labels, title, and more. Note that in matplotlib, the word axes is not the plural of axis: a plots axes contains much more information than just the x-axis and the y-axis. To know more about it, type "ax.plot?" in the next command line and execute.
